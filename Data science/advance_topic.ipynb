```python
time_df = pd.read_csv("./temprature.csv")
```

```python
import bar_chart_race as bcr
```

```python
time_df['Date'] = pd.to_datetime(time_df['Date'])
```

```python
arr = []
idx = []
i = 0
for value in time_df['Daily minimum temperatures']:
    try:
        arr.append(float(value))
    except Exception as error:
        idx.append(i)
        print(error)
    i+=1

print(arr)
print(idx)
```

```python
time_df = time_df.drop(index=idx)
```

```python
time_df['Daily minimum temperatures'] = time_df['Daily minimum temperatures'].astype(dtype=float)
```

```python
temp_df = pd.DataFrame()
temp_df['Daily minimum temperatures'] = time_df['Daily minimum temperatures']
```

```python
temp_df['Month'] = pd.DatetimeIndex(time_df['Date']).month
```

```python
bcr_df = bcr.load_dataset('covid19_tutorial')
```

```python
bcr.bar_chart_race(
        df=temp_df, 
        orientation='h', 
        filename='Temprature.mp4',
        sort='desc', 
        n_bars=12, 
        steps_per_period=20, 
        period_length=500, 
            ) 
```
